### Creating a balanced Sample

```python
real=real.sample(500)
fraud=fraud.sample(len(fraud))
real=real.sort_values("time")
fraud=fraud.sort_values("time")
balanced_data=pd.concat([real,fraud],axis=0).sample(len(real)+len(fraud))
```

### Data Spliting

```python
X=balanced_data.drop(columns="Class")
y=balanced_data["Class"]

X_full=data_cp.sample(len(data_cp))
y_full=X_full["Class"]
X_full=(X_full.drop(columns="Class"))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,train_size=0.8,test_size=0.2,shuffle=True)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from xgboost import XGBClassifier
```

```python
model_lr=LogisticRegression()
model_rfrc=RandomForestClassifier(n_estimators=400,max_depth=1,criterion="entropy")
model_dtc=DecisionTreeClassifier(max_depth=2)
model_abc=AdaBoostClassifier(n_estimators=100,learning_rate=0.1)
model_xgbc=XGBClassifier(max_depth=1,n_estimators=30)
models=[model_lr,
        model_rfrc,
        model_dtc,
        model_abc,
        model_xgbc]
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error,recall_score,f1_score
for m in models:
    m.fit(X_train,y_train)
    
    train_pred=m.predict(X_train)
    print('Train : ',accuracy_score(train_pred,y_train), mean_absolute_error(train_pred,y_train), recall_score(train_pred,y_train),f1_score(train_pred,y_train))
    
    valid_pred=m.predict(X_valid)
    print('Valid : ',accuracy_score(valid_pred,y_valid), mean_absolute_error(valid_pred,y_valid), recall_score(valid_pred,y_valid),f1_score(valid_pred,y_valid))
    
    full_pred=m.predict(X_full)
    print('Full : ',accuracy_score(full_pred,y_full), mean_absolute_error(full_pred,y_full), recall_score(full_pred,y_full), f1_score(full_pred,y_full),'\n')
```

```python
import joblib
joblib.dump(model_lr,"ccfd_LR")
joblib.dump(model_rfrc,"ccfd_RFRC")
joblib.dump(model_dtc,"ccfd_DTC")
joblib.dump(model_abc,"ccfd_ABC")
joblib.dump(model_xgbc,"ccfd_XGBC")
```
